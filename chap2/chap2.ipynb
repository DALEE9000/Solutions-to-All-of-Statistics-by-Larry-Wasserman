14. Let $(X,Y)$ be uniformly distributed on the unit disk $\{(x,y) : x^2 + y^2 \le 1\}$. Let $R = \sqrt{X^2 + Y^2}$. Find the CDF and PDF of $R$.

```python
R = 1
num_points = 1000000

# polar coordinates first, then convert to Cartesian
theta = np.random.uniform(0,2*np.pi, num_points)
radius = np.sqrt(np.random.uniform(0,R, num_points))

x = radius * np.cos(theta)
y = radius * np.sin(theta)
R = np.sqrt(np.square(x) + np.square(y))

plt.figure(figsize=(6,4))
plt.style.use(['science','scatter','dark_background'])
count, bins, ignored = plt.hist(R, 10000, density=True, color="#23895d")
plt.plot(bins, 2*bins, linewidth=2, color='#e2dd3f', zorder=1)
plt.text(0.3, 1.5, '$f_R(r) = 2r$', ha='center', va='center', fontsize=16)
plt.title('$\\textsf{PDF } f_R(r)$', fontsize=16)
plt.xlabel('$r$', fontsize=16)
plt.ylabel('$f_R(r)$', fontsize=16)
plt.minorticks_off()

plt.savefig('chap2ex14.eps', format='eps')
plt.show()
```

15. (A universal random number generator.) Let $X$ have a continuous, strictly increasing CDF $F$. Let $Y = F(X)$. Find the density of $Y$. This is called the probability integral transform. Now let $U \sim$ Uniform $(0,1)$ and let $X = F^{-1}(U)$. Show that $X \sim F$. Now write a program that takes Uniform $(0,1)$ random variables and generates random variables from an Exponential ($\beta$) distribution.

```python
def uniformgen():
    q = np.random.uniform(0,1,1)
    return q[0]

def expgen(q, beta):
    x = -beta*np.log(1-q)
    return x

beta = 0.4
expvals = []
for i in range(0,100000):
    expvals.append(expgen(uniformgen(),beta))

plt.figure(figsize=(6,4))
plt.style.use(['science','scatter','dark_background'])
count, bins, ignored = plt.hist(expvals, 10000, density=True, color="#44008b")
plt.plot(bins, (1/beta)*np.exp(-bins/beta), linewidth=2, color='#e54ed0', zorder=1)
plt.text(2, 1.5, '$f_X(x) = \\frac{1}{\\beta} e^{-x/\\beta}$', ha='center', va='center', fontsize=16)
plt.title('$\\textsf{PDF } f_X(x)$', fontsize=16)
plt.xlabel('$x$', fontsize=16)
plt.ylabel('$f_X(x)$', fontsize=16)
plt.minorticks_off()

plt.savefig('chap2ex15.eps', format='eps')
plt.show()
```

18. Let $X \sim N(3,16)$. Solve the following using the Normal table and using a computer package.

(a) Find $\mathbb{P}(X < 7)$.

(b) Find $\mathbb{P}(X > -2)$.

(c) Find $x$ such that $\mathbb{P}(X > x) = .05$.

(d) Find $\mathbb{P}(0 \le X < 4>)$.

(e) Find $x$ such that $\mathbb{P}(|X| > |x|) = .05$.

```python
from scipy.stats import norm

normal_dist = norm(3, 4)

# (a)
prob_a = normal_dist.cdf(7)
print("The probability for (a) is", prob_a)

# (b)
prob_b = 1 - normal_dist.cdf(-2)
print("The probability for (b) is", prob_b)

# (c)
x_c = normal_dist.ppf(0.95)
print("The value of x for (c) is", x_c)

# (d)
prob_d = normal_dist.cdf(4) - normal_dist.cdf(0)
print("The probability for (d) is", prob_d)

# (e)
x = np.arange(9.5,9.7,0.00000001)
array = np.round(normal_dist.cdf(x) - normal_dist.cdf(-x), decimals=9)
index = np.where(array == 0.95)
x_e = x[index[0]]
print("Some values of x for (e) are", x_e)
```
